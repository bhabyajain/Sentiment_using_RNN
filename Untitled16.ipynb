```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, GRU, Embedding
from tensorflow.python.keras.optimizers import Adam
from tensorflow.python.keras.preprocessing.text import Tokenizer
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
```

```python
train=pd.read_csv("twitter_train.csv")
test=pd.read_csv("twittter_test.csv")
```

```python
train
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['airline_sentiment'] = le.fit_transform(train['airline_sentiment'])
train['airline_sentiment']
```

```python
text=train["text"].values
pred=train["airline_sentiment"].values
text.shape , pred.shape
```

```python
from sklearn import model_selection
train_text,check_text,train_pred,check_pred=model_selection.train_test_split(text,pred,test_size=0.1)
```

```python
test_text=test["text"].values
```

```python
check_text
```

```python
num_words=10000
```

```python
tokenizer = Tokenizer(num_words=num_words)
```

```python
tokenizer.fit_on_texts(text)
```

```python
tokenizer.word_index
```

```python
train_tokens=tokenizer.texts_to_sequences(train_text)
```

```python
train_text[3]
```

```python
np.array(train_tokens[3])
```

```python
check_tokens=tokenizer.texts_to_sequences(check_text)
test_tokens=tokenizer.texts_to_sequences(test_text)
```

```python
num_tokens=[len(tokens) for tokens in train_tokens+check_tokens+test_tokens]
```

```python
np.mean(num_tokens) , np.max(num_tokens)
```

```python
pad='pre'
```

```python
train_pad = pad_sequences(train_tokens, maxlen=np.max(num_tokens),
                            padding=pad, truncating=pad)
```

```python
check_pad = pad_sequences(check_tokens, maxlen=np.max(num_tokens),
                            padding=pad, truncating=pad)
```

```python
test_pad = pad_sequences(test_tokens, maxlen=np.max(num_tokens),
                            padding=pad, truncating=pad)
```

```python
train_pad.shape , check_pad.shape, test_pad.shape
```

```python
train_pad[3]
```

```python
idx = tokenizer.word_index
inverse_map = dict(zip(idx.values(), idx.keys()))
def tokens_to_string(tokens):
    # Map from tokens back to words.
    words = [inverse_map[token] for token in tokens if token != 0]
    
    # Concatenate all words.
    text = " ".join(words)

    return text
```

```python
train_text[3]
```

```python
tokens_to_string(train_tokens[3])
```

```python
tokens_to_string(train_pad[3])
```

```python
model=Sequential()
```

```python
embedding_size=8
model.add(Embedding(input_dim=num_words,
                    output_dim=embedding_size,
                    input_length=np.max(num_tokens),
                    name='layer_embedding'))
```

```python
model.add(GRU(units=16, return_sequences=True))
```

```python
model.add(GRU(units=4))
```

```python
model.add(Dense(3, activation='sigmoid'))
```

```python
optimizer = Adam(lr=1e-3)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=optimizer,
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_pred_label_encoded=np.zeros((train_pred.shape[0],3))
for i in range(train_pred.shape[0]):
    train_pred_label_encoded[i][train_pred[i]]=1
```

```python
model.fit(train_pad, train_pred_label_encoded,
          validation_split=0.05, epochs=20,batch_size=64)
```
